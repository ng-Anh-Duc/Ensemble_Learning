```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/cleveland.csv')
df.head()
```

```python
df.info()
```

```python
dnull = df[df.isnull().any(axis=1)]
dnull
```

```python
for i in range(df.shape[0]):
  if pd.isnull(df.loc[i,'ca']):
    cvalue = df.loc[i,'target']
    df.loc[i,'ca'] = df[df['target'] == cvalue].mode().iloc[0,11]
```

```python
df.tail()
```

```python
df = df.dropna()
df
```

```python
import numpy as np

X = df.iloc[:,:-1]
y = df.iloc[:,-1]
X = np.array(X)
X
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,stratify=y, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import StackingClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
classifier = KNeighborsClassifier(n_neighbors=5, algorithm='kd_tree')
classifier.fit(X_train, y_train).score(X_test, y_test)
```

```python
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train).score(X_test, y_test)
```

```python
# n_neighbors=5, weights='distance', algorithm='auto'
estimators = [('knn', KNeighborsClassifier()),
              ('nb', GaussianNB()),
              ('dt', DecisionTreeClassifier(random_state=42)),
              ('rf', RandomForestClassifier(random_state=42)),
              ('ada', AdaBoostClassifier(random_state=42)),
              ('svc', SVC(kernel='rbf', random_state=42)),
              ('gb', GradientBoostingClassifier(random_state=42))]

clf = StackingClassifier(estimators=estimators, final_estimator=LogisticRegression())
clf.fit(X_train, y_train)
```

```python
y_preds = clf.predict(X_test)
print(confusion_matrix(y_preds, y_test))
print(accuracy_score(y_test, y_preds))
print(y_preds)
```
